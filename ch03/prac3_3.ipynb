```python
import itertools
def rank_corr(df):
    n = len(df.index)
    for clm in itertools.combinations(df.columns, 2):
        diff2 = (df[clm[0]] - df[clm[1]])**2
        r = 1 - 6/(n**3 - n) * diff2.sum()
        print(clm[0]+"と"+clm[1]+"の順位相関係数 : "+str(r))
    
rank_corr(df)
```
