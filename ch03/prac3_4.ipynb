```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
df = pd.DataFrame(
    {'men(brothers)' : [71, 68, 66, 67, 70, 71, 70, 73, 72, 65, 66],
     'women(sisters)' : [69, 64, 65, 63, 65, 62, 65, 64, 66, 59, 62]
    },
    columns=['men(brothers)', 'women(sisters)']
    )
df.plot.scatter(x='men(brothers)', y='women(sisters)')
```

```python
rand_indx = np.random.randint(0,12,(200,11))
corr_coeffs = np.array([df.ix[i,:].corr().values[0,1] for i in rand_indx])
plt.hist(corr_coeffs)
plt.xlabel("Correlation coefficient [a.u.]")
plt.ylabel("Frequency (n=200) [a.u.]")
plt.show()
```
